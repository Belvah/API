```python
#\d is a regular expression pattern that matches any digit character (0-9). It is a shorthand character class that represents any single digit. For example, if you use the pattern r"\d" in a regular expression search, it will match any digit in the input string.
#\D is a regular expression pattern that matches any non-digit character. It is the opposite of \d, which matches digits. For example, if you use the pattern r"\D" in a regular expression search, it will match any character that is not a digit in the input string.

import re

text = "My Project A: file_20, My Project B: file_35, Project C: file_7, Project D: file_786"

match = re.findall(r"file_\d\d", text) #the regular expression pattern r"file_\d\d" is used to find all occurrences of the string "file_" followed by two digits in the given text. The \d represents a digit character, and since there are two \d in the pattern, it will match any string that starts with "file_" and is followed by exactly two digits.
print(match)
```

```python
import re

html = "<b>Bold</b> <i>Italic</i>"

# GREEDY (The Default)
# Matches from the first < to the LAST >
greedy = re.findall(r"<.+>", html)
print(greedy)
# Output: ['<b>Bold</b> <i>Italic</i>'] (One giant string!)

# LAZY (The Fix)
# Matches from < to the FIRST >
lazy = re.findall(r"<.+?>", html)
print(lazy)
# Output: ['<b>', '</b>', '<i>', '</i>'] (Distinct tags)
```

```python

#Goal: Write Regex code that is maintainable (easy to read), robust (handles messy real-world data), and efficient (doesn't crash your memory).


import re

log_entry = "Event: LOGIN_SUCCESS User: admin_01 Time: 14:30"

# FRAGILE WAY (Numbered Groups)
# If we add anything to the start of this pattern, indexes shift.
fragile_pattern = r"Event: (\w+) User: (\w+) Time: (\d+:\d+)"

# PROFESSIONAL WAY (Named Groups)
# We embed the keys 'event', 'user', and 'time' into the pattern.
robust_pattern = r"Event: (?P<event>\w+) User: (?P<user>\w+) Time: (?P<time>\d+:\d+)"

match = re.search(robust_pattern, log_entry)

if match:
    # 1. Access by Name (Readable)
    print(f"User: {match.group('user')}")
    print(f"Action: {match.group('event')}")
    
    # 2. Convert to Dictionary (Powerful)
    # This creates: {'event': 'LOGIN_SUCCESS', 'user': 'admin_01', 'time': '14:30'}
    data = match.groupdict() 
    print(data['time'])


```

```python
#FLAGS-COMBINING FLAGS.

import re

text = """
Start
Title: PYTHON 
Description: Python is a
versatile language.
End
"""

# Scenario: Extract everything between 'Title:' and 'End', regardless of case or newlines.

# 1. Case Insensitive (Matches 'PYTHON' even though we wrote 'python')
# 2. Dot All (Matches the newlines inside the description)
pattern = r"title: (.*?)\s+end"
#added \s+ before end to match any whitespaces including new line.

match = re.search(pattern, text, re.IGNORECASE | re.DOTALL)

if match:
    print(match.group(1).strip())
  
# Output:
# PYTHON 
# Description: Python is a
# versatile language.
```

```python
#Advanced: Using Backreferences in Replacement

#Scenario: Changing Date Format Convert "MM/DD/YYYY" (US format) to "YYYY-MM-DD" (ISO format).

date_list = "Dates: 12/31/2023 and 01/15/2024"

# Step 1: Capture the parts (Month)/(Day)/(Year)
pattern = r"(\d{2})/(\d{2})/(\d{4})"

# Step 2: Re-arrange them using \3, \1, \2
# \3 refers to the 3rd group (Year)
# \1 refers to the 1st group (Month)
# \2 refers to the 2nd group (Day)
reformatted = re.sub(pattern, r"\3-\1-\2", date_list)

print(reformatted)
# Output: "Dates: 2023-12-31 and 2024-01-15"

```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

#Initialize the Container (The Empty List)
#This MUST be outside the loop, or you will overwrite your data every time.
books_data = []


url = "http://books.toscrape.com/catalogue/page-3.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")


cards = soup.select(".product_pod")

for card in cards:
    # Extract the raw data
    title = card.select_one("h3 a")["title"]
    price_text = card.select_one(".price_color").text
    stock = card.select_one(".availability").get_text(strip=True)
    
    # 3. Create the Dictionary (The Row)
    # The Keys you choose here ('Title', 'Cost') will become your CSV headers.
    book_item = {
        "Title": title,
        "Cost": price_text,
        "Stock Status": stock,
        "Source_URL": url  # Good practice: Track where data came from
    }
    
    # Append the row to the main list
    books_data.append(book_item)
```

```python
WEB SERVICES.

#web services are software systems designed to support interoperable machine-to-machine interaction over a network. They allow different applications to communicate with each other and share data and services. Web services typically use standard protocols such as HTTP, XML, SOAP, and REST to enable communication between different systems. They can be used for various purposes, including data exchange, remote procedure calls, and integration of different applications. Web services play a crucial role in enabling the development of distributed applications and facilitating communication between different software components.

#tight coupling is when a system fails to separate concerns, making it difficult to modify or maintain one part of the system without affecting other parts. In a tightly coupled system, components are highly dependent on each other, which can lead to issues such as increased complexity, reduced flexibility, and difficulty in testing and debugging. Tight coupling can make it challenging to adapt to changes or add new features without causing unintended consequences in other parts of the system.

#loose coupling, on the other hand, is when a system is designed to minimize dependencies between components. In a loosely coupled system, components are independent and can be modified or maintained without affecting other parts of the system. Loose coupling promotes flexibility, scalability, and maintainability, as changes to one component do not have a significant impact on others. It allows for easier testing and debugging, as well as the ability to adapt to changes or add new features without causing unintended consequences in other parts of the system.


4 pillars of web services

*service provider-specific endpoints defined.
*service requester e.g mobile app / anything requesting for data.
*endpoint-specific url service accessed.
*payload- search terms, data, etc. that is sent to the service provider for processing and retrieval of information.

The request-response cycle of web services involves the following steps:
>formulation-client gathers data(the service requester)
>Transmission(request)-use HTTP methods e.g Get n provide endpoint.
>Processing-service provider processes the request.
>Transmission(Response)service provider sends back the requested data or result.
>Consumption.

Types of web services:
1. SOAP (Simple Object Access Protocol): A protocol that uses XML for message formatting and relies on other application layer protocols, such as HTTP or SMTP, for message transmission. 
2. REST (Representational State Transfer): An architectural style that uses standard HTTP methods and relies on stateless communication. RESTful web services typically use JSON or XML for data exchange.
3. XML-RPC (Remote Procedure Call): A protocol that uses XML to encode remote procedure

Benefits of web services:
1. Interoperability- Making systems work together regardless of their underlying technologies.
2. Reusability- Web services can be reused across different applications and platforms, reducing development time and effort.
3. Scalability- Web services can handle a large number of requests and can be easily scaled to accommodate growing user demands.
4. Flexibility- Web services can be accessed from various devices and platforms, allowing for greater flexibility in application development and deployment.
5. Integration- Web services enable seamless integration between different applications and systems, facilitating data exchange and collaboration.
6. Cost-effectiveness- Web services can reduce development costs by allowing developers to leverage existing services and components, rather than building everything from scratch.


XML (eXtensible Markup Language) is a markup language that defines a set of rules for encoding documents in a format that is both human-readable and machine-readable. It is commonly used for data exchange and storage, as well as for defining the structure of web services. XML allows for the creation of custom tags and attributes, making it flexible and adaptable to various applications.   

STRUCTURE.

<root>
<child>
<element></element>
</child>
</root>

#values of attributes in speechmarks
```

```python
#OBJECT ORIENTED PROGRAMMING (OOP) IN PYTHON.

class BankAccount:
    def __init__(self, name, balance): #condtructor method.
        # init is a special method that is called when an object is created from a class.
        # It is used to initialize the attributes of the object.
        # The self parameter refers to the instance of the class being created.
        self.name = name
        self.balance = balance

    def withdraw(self,amount):
        if amount >self.balance:
            print("You have insufficient funds in your account.Try a lower amount to withdraw.")
        else:
            self.balance -= amount

Account1=BankAccount("Grace",10000)
Account2=BankAccount("Belvah", 100000)
Account3=BankAccount("Nancy", 150000)
Account1.withdraw(1000)  #for Grace to withdraw         
```
