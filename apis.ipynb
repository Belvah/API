```python
import requests 
import csv

url = 'https://jsonplaceholder.typicode.com/users/1'

response = requests.get(url) #get request is for reading data from the api. The response variable will store the response from the api after making a get request to the url.

print(f"The api responded with:{response.status_code}")
if response.status_code == 200:
    data = response.json() #change the response to json format and store it in a variable called data.
    print(data['name'])
    print(data['username'])
    print(data['address'])    
    
    #finish up the data fetching by having the data written in a text file.Open a txt file in write mode and write to it the data that you get from the api.

    file = open('new_file.txt', 'w')
    file.write(f"Name: {data['name']}\n")
    file.write(f"Username: {data['username']}\n")
    file.write(f"Address: {data['address']}\n")
    file.close()
    
    print("This is a new file.txt")
```

```python
#POST REQUEST for creating data.

import requests

base_url = 'https://jsonplaceholder.typicode.com/posts'

new_payload = {
    "title": "Mastering python APIs",
    "body": "Get a detailed guide on how to work with APIs in python",
    "userId": 99
}

response = requests.post(base_url, json=new_payload) #response is the variable that will store the response from the api after making a post request to the base url with the new payload as the data to be sent to the api.

if response.status_code == 201:
    print("Post request successful", response.json()['id'])
```

```python
#PUT REQUEST for updating data.
import requests

base_url = 'https://jsonplaceholder.typicode.com/posts/1'

updated_payload = {
    'id': 1,
    "title": "Mastering python regular expressions",
    "userID" : 1
   
}

response = requests.put(base_url, json=updated_payload) #response is the variable that will store the response from the api after making a put request to the base url with the updated payload as the data to be sent to the api.
data = response.json()
if response.status_code == 200:
    print("Data updated successfully", data)
    print(response)
```

```python
#DELETE REQUEST for deleting data.
from requests.exceptions import HTTPError, ConnectionError, Timeout  

base_url = 'https://jsonplaceholder.typicode.com/posts/1'

try:
    response = requests.delete(base_url)
    
    if response.status_code == 204:
        print("Data deleted successfully")
except Timeout:
    print("Server took too long to respond")
except HTTPError as err:
    print(f"Unexpected error occurred: {err}")
    print(response)
    
      
```

```python
import requests
import csv

#PENDING: CREATE A CSV file.

url = 'https://jsonplaceholder.typicode.com/users'

response = requests.get(url) #get request is for reading data from the api. The response variable will store the response from the api after making a get request to the url.

print(f"The api responded with:{response.status_code}")
if response.status_code == 200:
    data = response.json() #change the response to json format and store it in a variable called data.
    
    with open('new_file.txt', 'w') as file:
        writer = csv.writer(file)
        writer.writerow(['ID', 'Name', 'Email', 'City', 'Latitude', 'Longitude'])   

        for user in data:
            row = [
                user['id'],
                user['name'],
                user['email'],
                user['address']['city'],
                user['address']['geo']['lat'],
                user['address']['geo']['lng'],
            ]
            writer.writerow(row)
    
    print('data printed successfully to csv file')
```

```python
#Regular expressions are a powerful tool for pattern matching and text manipulation. They allow you to search for specific patterns in strings, extract information, and perform various operations on text data. In Python, the `re` module provides functions for working with regular expressions.

import re

pattern ='path\to\new\folder'
match = re.search(r'path\to\new\folder', pattern)#pass r before the pattern to indicate that it is a raw string, which allows you to use backslashes without needing to escape them.

print(match)




```

```python
#Basic search for a pattern in a string.

#data = "Order ID: 2456456"
data = "Oder ID: 2456456"

match = re.search(r"Order" , data) #search for the pattern "Order" in the string stored in the variable data. The re.search() function returns a match object if the pattern is found, or None if it is not found.

if match: #if the match variable is not None, it means that the pattern was found in the string, and we can print a message indicating that the pattern was found.
    print("Pattern found!")
    print("The object is:" , match)
else:
    print("Pattern not found.")
```

```python
import re #import the re module, which provides functions for working with regular expressions in Python.

text = "The sky was Gray, the shirt was Grey, the name of alien was Graay"

match = re.findall(r"Gr.y", text) #"Gr.y" is a regular expression pattern that matches any string that starts with "Gr", followed by any single character (represented by the dot), and ends with "y". The re.findall() function returns a list of all non-overlapping matches of the pattern in the string.

print(match)
```

```python
#\d is a regular expression pattern that matches any digit character (0-9). It is a shorthand character class that represents any single digit. For example, if you use the pattern r"\d" in a regular expression search, it will match any digit in the input string.
#\D is a regular expression pattern that matches any non-digit character. It is the opposite of \d, which matches digits. For example, if you use the pattern r"\D" in a regular expression search, it will match any character that is not a digit in the input string.

import re

text = "My Project A: file_20, My Project B: file_35, Project C: file_7, Project D: file_786"

match = re.findall(r"file_\d\d", text) #the regular expression pattern r"file_\d\d" is used to find all occurrences of the string "file_" followed by two digits in the given text. The \d represents a digit character, and since there are two \d in the pattern, it will match any string that starts with "file_" and is followed by exactly two digits.
print(match)
```

```python

#\W is a regular expression pattern that matches any non-word character. It is the opposite of \w, which matches word characters (letters, digits, and underscores). For example, if you use the pattern r"\W" in a regular expression search, it will match any character that is not a letter, digit, or underscore in the input string.
import re

text = "My Project A: file_20, My Project B: file_35, Project C: file_7, Project D: file_786"

match = re.findall(r"\w", text) #returns all non-word characters in the given text. The \W is a shorthand character class that matches any character that is not a letter, digit, or underscore. The re.findall() function will return a list of all the characters that match this pattern in the input string.
print(match)
```

```python
#$ is a regular expression character that matches the end of a string. It is used to indicate that the pattern should be matched at the end of the string. For example, if you use the pattern r"Error$" in a regular expression search, it will match any string that ends with "Error". The re.search() function returns a match object if the pattern is found, or None if it is not found.

import re

file = 'logo.png'

if re.search(r"png$", file): #search for the pattern "png" at the end of the string stored in the variable file. The $ character is used to indicate that the pattern should be matched at the end of the string. The re.search() function returns a match object if the pattern is found, or None if it is not found.
    print("This is a png file!")
else:
    print("No such file found.")
```

```python
#\b is used to set boundaries i.e from what point to what point should the pattern be matched. For example, if you use the pattern r"\bcat\b" in a regular expression search, it will match the word "cat" only if it is a standalone word, and not part of another word like "catalog" or "scatter". The \b represents a word boundary, which is the position between a word character and a non-word character.Returns specific words.

import re

text = 'My favourite pet is Loki, she is a gorgeous pet!'

print(re.findall(r"\bLoki\b" , text))

```

```python
import re

error_codes = 'Errors: 404, 401, 500 phone: 254-567-890, color:colour'

print(re.findall(r'\d{3}' , error_codes)) #the regular expression pattern r'\d{3}' is used to find all occurrences of exactly three digits in the given string error_codes. The \d represents a digit character, and the {3} indicates that there should be exactly three of them. The re.findall() function returns a list of all non-overlapping matches of the pattern in the string. In this case, it will return a list of all the three-digit numbers found in the error_codes string.

print(re.findall(r'\d+' , error_codes)) #the regular expression pattern r'\d+' is used to find all occurrences of one or more digits in the given string error_codes. The \d represents a digit character, and the + indicates that there can be one or more of them. The re.findall() function returns a list of all non-overlapping matches of the pattern in the string. In this case, it will return a list of all the numbers found in the error_codes string.

print(re.findall(r'\d{3}-\d{3}-\d{3}' , error_codes)) #the regular expression pattern r'\d{3}-\d{3}-\d{3}' is used to find all occurrences of a phone number pattern in the given string error_codes. The \d represents a digit character, and the {3} indicates that there should be exactly three of them. The - is used to match the hyphen character. The re.findall() function returns a list of all non-overlapping matches of the pattern in the string. In this case, it will return a list of all the phone numbers found in the error_codes string.

print(re.findall(r'colour' , error_codes)) #the regular expression pattern r'colour' is used to find all occurrences of the word 'colour' in the given string error_codes. The re.findall() function returns a list of all non-overlapping matches of the pattern in the string. In this case, it will return a list containing the word 'colour' if it is found in the error_codes string.
```

```python

import re

log = 'Time: 21:11 Error:Error'

print(re.findall(r'Time=\d+:\d+', log)) #the regular expression pattern r'Time=\d+:\d+' is used to find all occurrences of a time pattern in the given string log. The Time= is a literal string that matches the characters "Time=". The \d represents a digit character, and the + indicates that there can be one or more of them. The : is used to match the colon character. The re.findall() function returns a list of all non-overlapping matches of the pattern in the string. In this case, it will return a list containing the time found in the log string if it matches the specified pattern.

#Lets clean up the above.

print(re.findall(r'Time = (\d+:\d+)', log)) #the regular expression pattern r'Time = (\d+:\d+)' is used to find all occurrences of a time pattern in the given string log, but with a slight modification. The Time = is a literal string that matches the characters "Time =". The parentheses around \d+:\d+ create a capturing group, which allows us to extract just the time portion of the match. The \d represents a digit character, and the + indicates that there can be one or more of them. The : is used to match the colon character. The re.findall() function returns a list of all non-overlapping matches of the pattern in the string. In this case, it will return a list containing only the time found in the log string if it matches the specified pattern, without including the "Time = " part.
```
